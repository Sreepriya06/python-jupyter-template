### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Data for plots
x = np.arange(1, 6)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'purple']

# Line plot (Top-left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X Values")
axs[0, 0].set_ylabel("Y Values")
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Bar plot (Top-right)
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter plot (Bottom-left)
axs[1, 0].scatter(x, scatter_y, color='blue', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X Values")
axs[1, 0].set_ylabel("Y Values")
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Pie chart (Bottom-right)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title("Pie Chart")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
